```python
#encoding the categorical target variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
column = ['Species']
data[column] = data[column].apply(le.fit_transform)
data
```

# Data modelling

```python
X = data.drop(['Species'],axis='columns')
```

```python
y = pd.DataFrame(y)
print(y)
type(y)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

# Model Selection

```python
# Using Logistic Regression
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
pred
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
print("Accuracy of the model :",accuracy_score(y_test,pred)*100, '%')
```

```python
#Predicting flower species
SepalLength = float(input("Enter SepalLength in cm :"))
SepalWidth = float(input("Enter SepalWidth in cm :"))
PetalLength = float(input("Enter PetalLength in cm :"))
PetalWidth = float(input("Enter PetalWidth in cm :"))
df = [[SepalLength,SepalWidth,PetalLength,PetalWidth]]
result = reg.predict(df)
print(result)

if result == 0:
    print("Iris-setosa")
elif result == 1:
     print("Iris-versicolor") 
else:
    print("Iris-virginica")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
```

```python
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
acc = dt.score(X_test,y_test)
acc
```

```python
#using svm
from sklearn.svm import SVC
svc = SVC (kernel='linear', random_state=0)
svc.fit(X_train, y_train)
```

```python
y_prediction = svc.predict(X_test)
y_prediction
```

```python
acc = accuracy_score(y_test,y_prediction)
acc
```

```python
# classification reports
from sklearn.metrics import classification_report
print(classification_report(y_test,y_prediction))
```

```python
# confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
cm = confusion_matrix(y_test,y_prediction)
cm
```

The Iris dataset is quite small with only 150 observations.  It's observations are also evenly divided into three classes which are iris-setosa, iris-virginica and iris-versicolor. This analysis aims to classify iris flowers among three species (iris-setosa, iris-virginica, iris-versicolor) based on the length and width measurements of sepals and petals. Here the model makes proper classifications for new flowers and got an accuracy of 97.77%, which shows that the model is very accurate.
